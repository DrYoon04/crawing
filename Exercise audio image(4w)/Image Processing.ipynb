```python
from PIL import Image
```

```python
img = Image.open('cat-1151519.jpg')
img.show()
```

```python
print(img.filename)
print(img.format)
print(img.size)
print(img.width)
print(img.height)
print(img.mode)
```

```python
img_resize = img.resize((300, 300))
img_resize.show()

img.show()
```

```python
img_crop = img.crop((100, 100, 400, 400))

img_crop.show()
```

```python
img_rotate = img.rotate(45)

img_rotate.show()
```

```python
img_flipped_LR = img.transpose(Image.FLIP_LEFT_RIGHT)
img_flipped_LR.show()

img_flipped_TB = img.transpose(Image.FLIP_TOP_BOTTOM)
img_flipped_TB.show()
```

```python
img_flipped_TB.save('cat-1151519-flipped.jpg')
```

```python
img_gray = img.convert('L')
img_gray.show()
```

```python
from PIL import ImageFilter
```

```python

img_blur = img.filter(ImageFilter.GaussianBlur(15))

img_blur.show()
```

```python
img_edge = img.filter(ImageFilter.EDGE_ENHANCE)
img_edge.show()
```

```python
import os , sys
from PIL import Image
```

```python
size = (128, 128)

for infile in sys.argv:
    outfile = os.path.splitext(infile)[0] + ".thumbnail"
    if infile != outfile:
        try:
            im = Image.open(infile)
            im.thumbnail(size)
            im.save(outfile, "JPEG")
        except IOError:
            print("cannot create thumbnail for", infile)
```
